# Data Preprocessing

```python
import pandas as pd

df = pd.read_csv("../data/amazon_data.csv")

categorical_columns = df.select_dtypes(include=["object"]).columns
numerical_columns = df.select_dtypes(include=["number"]).columns

for col in numerical_columns:
    df[col] = df[col].fillna(df[col].mean()).astype(int)

df.to_csv("cleaned_dataset.csv", index=False)

print("Data cleaning completed and saved as cleaned_dataset.csv")
```

# Image PreProcessing

```python
import cv2
import os
import numpy as np

input_folder = "../data/product_images/"  
output_folder = "processed_images/"  
os.makedirs(output_folder, exist_ok=True)

IMG_SIZE = (224, 224)

for filename in os.listdir(input_folder):
    if filename.endswith(('.jpg', '.png', '.jpeg')):  
        img_path = os.path.join(input_folder, filename)
        img = cv2.imread(img_path)

        if img is None:
            print(f"Skipping {filename}: Unable to load image.")
            continue
        
        img_resized = cv2.resize(img, IMG_SIZE)

        save_path = os.path.join(output_folder, filename)
        cv2.imwrite(save_path, img_resized)  
        print(f"Processed: {filename}")

print("✅ Image preprocessing completed!")
```
